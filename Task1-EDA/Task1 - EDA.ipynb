## Hot encoding

```python
df['gender'].replace(["male","female"],[0,1],inplace=True)
```

```python
df['race/ethnicity'].replace(["group A","group B","group C","group D","group E"],[0,1,2,3,4],inplace=True)
```

```python
df['parental level of education'].replace(["high school","some college","associate's degree","bachelor's degree","master's degree"],[0,1,2,3,4],inplace=True)
```

```python
df['lunch'].replace(["standard","free/reduced"],[0,1],inplace=True)
```

```python
df['test preparation course'].replace(["none","completed"],[0,1],inplace=True)
```

## Data Dictionary:
##### Gender : 0 = Male, 1= Female
##### Race/Ethnicity :  0 = group A, 1 = group B, 2 = group C, 3 = group D, 4 = group E 
#####  Parental level of education : 0 = high school/some high school, 1 = some college, 2 = associates degree, 3 = bachelors degree, 4 = masters degree
##### Lunch : 0 = standard, 1 = free/reduced
##### Test preparation course : 0 = none, 1 = completed
